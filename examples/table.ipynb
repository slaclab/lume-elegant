# Parse SDDS table

```python
from elegant.parsers import parse_sdds_table
```

```python
?parse_sdds_table
```

```python
dat = parse_sdds_table('data/LCLS2scH.twi', ['s', 'betax', 'betay', 'etax'])
dat
```

## Plot

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(dat['s'], dat['betax'], label=r'$\beta_x$')
ax.plot(dat['s'], dat['betay'], label=r'$\beta_y$')
plt.legend()
ax.set_xlabel('s (m)')
ax.set_ylabel(r'$\beta_{x,y}$ (m)')
```
